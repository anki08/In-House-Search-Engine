```python
def get_sentence_embeding(sentences):
  preprocessed_text = sbert_model.encode(sentences)
  return preprocessed_text


df_quest['encoded_questions']= df['question'].apply(lambda x:get_sentence_embeding([x]))
```

```python
#check for all questions
# df_test = df_quest.head("10")
from sklearn.metrics.pairwise import cosine_similarity
similar_vector_values=[]
predicted_ans_list=[]
max_cosine_list=[]
for index, row in df_quest.iterrows():
  query=row['question']
  #print(query)
  query_vectors = sbert_model.encode([query]) 
  for index, row in df.iterrows():
      ans = row['encoded_questions']
      similar_vector_values.append(np.sum(cosine_similarity(ans,query_vectors)))

  #print(similar_vector_values)
  dx = np.argmax(np.array(similar_vector_values))
  #print(dx)
  max_cosine=np.max(np.array(similar_vector_values))
  #print(max_cosine)

  similar_vector_values=[]
  predicted_ans=df.iloc[dx]['answer']
  predicted_ans_list.append(predicted_ans)
  max_cosine_list.append(max_cosine)

print("here", predicted_ans_list)
print(max_cosine_list)

```

```python
df_quest['predicted_answers']=predicted_ans_list
```

```python
df_quest['cosine_simalrity']=max_cosine_list
```

```python
new_df = pd.read_csv("/content/drive/MyDrive/covid_exttra_questions.csv")
new_df.head(5)
```

```python
#check for new questions
# df_test = df_quest.head("10")
from sklearn.metrics.pairwise import cosine_similarity
similar_vector_values=[]
predicted_ans_list=[]
max_cosine_list=[]
predicted_question_list=[]
for index, row in new_df.iterrows():
  query=row['What is a novel coronavirus?']
  #print(query)
  query_vectors = sbert_model.encode([query]) 
  for index, row in df.iterrows():
      ans = row['encoded_questions']
      similar_vector_values.append(np.sum(cosine_similarity(ans,query_vectors)))

  #print(similar_vector_values)
  dx = np.argmax(np.array(similar_vector_values))
  #print(dx)
  max_cosine=np.max(np.array(similar_vector_values))
  #print(max_cosine)

  similar_vector_values=[]
  predicted_ans=df.iloc[dx]['answer']
  predcited_question=df.iloc[dx]['question']
  predicted_ans_list.append(predicted_ans)
  max_cosine_list.append(max_cosine)
  predicted_question_list.append(predcited_question)

print("here", predicted_ans_list)
print(max_cosine_list)
print(predicted_question_list)
```
